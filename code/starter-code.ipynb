12. Write a function using only list comprehensions, no loops, to compute standard deviation. Using this function, calculate the standard deviation of each numeric column in both data sets. Add these to a list called `sd`.

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

```python
import numpy as np
def find_std(any_list):
   mean=np.mean(any_list)
   st=np.sqrt(sum([(i-mean)**2 for i in any_list])*1/len(any_list))
   return st
find_std(sat['Math'])
```

```python
sd=[find_std(sat['Evidence-Based Reading and Writing']),find_std(sat['Total']),find_std(sat['Math']),
   find_std(act['English']),find_std(act['Math']),find_std(act['Reading']),find_std(act['Science'])
   ,find_std(act['Composite'])]
sd
```

##### 26. Execute a hypothesis test comparing the SAT and ACT participation rates. Use $\alpha = 0.05$. Be sure to interpret your results.

```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

H0: The mean difference between SAT Participation and ACT Particiaption is zero.
H1: The mean difference between SAT Participation and ACT Particiaption  is different than zero.

```python
Participation_act_mean = np.mean(merged_doc['Participation_act'])
Participation_act_sem = np.std(merged_doc['Participation_act'])/np.sqrt(len(merged_doc['Participation_act']))

Participation_sat_mean = np.mean(merged_doc['Participation_sat'])
Participation_sat_sem = np.std(merged_doc['Participation_sat'])/np.sqrt(len(merged_doc['Participation_sat']))
```

```python

result = stats.ttest_ind(merged_doc['Participation_sat'], merged_doc['Participation_act'])
result.statistic, result.pvalue
```

```python
t_stat, p_value = stats.ttest_ind(merged_doc['Participation_sat'], merged_doc['Participation_act'])
```

The p value is 0.00024134203698662353 and much less than 0.05 thus we reject Ho hypothesis, thus there is statistical difference between two groups. The participation rate might be different for these two tests.

##### 27. Generate and interpret 95% confidence intervals for SAT and ACT participation rates.

```python
#95% CI for Participation_sat
stats.norm.interval(0.95, Participation_sat_mean, Participation_sat_sem)
```

```python
#95% CI for Participation_act
stats.norm.interval(0.95, Participation_act_mean, Participation_act_sem)
```
